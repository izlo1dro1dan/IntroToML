# The Tipping Problem
Taking the tipping example full circle, if we were to create a controller which estimates the tip we should give at a restaurant, we might structure it as such:

## Antecednets (Inputs)
* service
    * Universe (ie, crisp value range): How good was the service of the waitress, on a scale of 1 to 10?
    * Fuzzy set (ie, fuzzy value range): poor, acceptable, amazing
* food quality
    * Universe: How tasty was the food, on a scale of 1 to 10?
    * Fuzzy set: bad, decent, great

## Consequents (Outputs)
* tip
    * Universe: How much should we tip, on a scale of 0% to 25%
    * Fuzzy set: low, medium, high
* Rules
    * IF the service was good or the food quality was good, THEN the tip will be high.
    * IF the service was average, THEN the tip will be medium.
    * IF the service was poor and the food quality was poor THEN the tip will be low.
* Usage
    * If I tell this controller that I rated:
        * the service as 9.8, and
        * the quality as 6.5,
    * it would recommend I leave:
        * a 20.2% tip.

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import ipywidgets as widgets
from ipywidgets import interact_manual
# You can see how these look with .view()
%matplotlib inline

# New Antecedent/Consequent objects hold universe variables and membership
# functions
service = ctrl.Antecedent(np.arange(0, 11, 1), 'service')
quality = ctrl.Antecedent(np.arange(0, 11, 1), 'quality')
tip = ctrl.Consequent(np.arange(0, 26, 1), 'tip')
 
quality.graph
    
# Auto-membership function population is possible with .automf(3, 5, or 7)
service.automf(3)
quality.automf(3)

service.terms.pop('bad', None)
service.terms.pop('average', None)
service.terms.pop('good', None)
service['poor'] = fuzz.trimf(service.universe, [0, 0, 5])
service['acceptable'] = fuzz.trimf(service.universe, [0, 5, 10])
service['amazing'] = fuzz.trimf(service.universe, [5, 10, 10])
service.view()

quality.terms.pop('poor', None)
quality.terms.pop('average', None)
quality.terms.pop('good', None)
quality['bad'] = fuzz.trimf(quality.universe, [0, 0, 5])
quality['decent'] = fuzz.trimf(quality.universe, [0, 5, 10])
quality['great'] = fuzz.trimf(quality.universe, [5, 10, 10])
quality.view()

tip['low'] = fuzz.trimf(tip.universe, [0, 0, 13])
tip['medium'] = fuzz.trimf(tip.universe, [0, 13, 25])
tip['high'] = fuzz.trimf(tip.universe, [13, 25, 25])
tip.view()

rule1 = ctrl.Rule(service['poor'] | quality['bad'], tip['low'])
rule2 = ctrl.Rule(service['acceptable'], tip['medium'])
rule3 = ctrl.Rule(service['amazing'] | quality['great'], tip['high'])

tipping_ctrl = ctrl.ControlSystem([rule1, rule2, rule3])
tipping = ctrl.ControlSystemSimulation(tipping_ctrl)

def callInput(tipping_par):
    def get_input(srv, qlt):
        tipping_par.input['service'] = srv
        tipping_par.input['quality'] = qlt
    
        tipping_par.compute()
    
        print('Recommended tip', tipping_par.output['tip'])
        tip.view(sim=tipping_par)

    interact_manual(get_input, srv = widgets.FloatSlider(min = 0.0, max = 10.0, step = 0.1, value = 9.8), qlt = widgets.FloatSlider(min = 0.0, max = 10.0, step = 0.1, value = 6.5))

callInput(tipping)
```

```python
# Custom membership functions can be built interactively with a familiar,
# Pythonic API
tip.terms.pop('low', None)
tip.terms.pop('medium', None)
tip.terms.pop('high', None)
tip['poor'] = fuzz.trimf(tip.universe, [0, 0, 5])
tip['bad'] = fuzz.trimf(tip.universe, [0, 5, 10])
tip['average'] = fuzz.trimf(tip.universe, [5, 10, 15])
tip['descent'] = fuzz.trimf(tip.universe, [10, 15, 20])
tip['great'] = fuzz.trimf(tip.universe, [20, 25, 25])

rule1 = ctrl.Rule(quality['bad'] | service['poor'], tip['poor'])
rule2 = ctrl.Rule(quality['bad'] | service['acceptable'], tip['bad'])
rule3 = ctrl.Rule(quality['decent'] | service['poor'], tip['bad'])
rule4 = ctrl.Rule(quality['decent'] | service['acceptable'], tip['average'])
rule5 = ctrl.Rule(quality['decent'] | service['amazing'], tip['descent'])
rule6 = ctrl.Rule(quality['great'] | service['acceptable'], tip['descent'])
rule7 = ctrl.Rule(quality['great'] | service['amazing'], tip['great'])

tipping_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7])
tipping = ctrl.ControlSystemSimulation(tipping_ctrl)

callInput(tipping)
```

```python
from ipywidgets import interact_manual

#Added new criteria of evaluation (interior and quota)
interior = ctrl.Antecedent(np.arange(0, 11, 1), 'interior')
quota = ctrl.Antecedent(np.arange(0, 11, 1), 'quota')

interior.automf(3)
quota.automf(3)

rule1 = ctrl.Rule(quality['bad'] | service['poor'], tip['poor'])
rule2 = ctrl.Rule(interior['poor'] | quota['good'], tip['bad'])
rule3 = ctrl.Rule(service['amazing'] | quality['decent'], tip['descent'])
rule4 = ctrl.Rule(service['amazing'] | quality['great'], tip['great'])
rule5 = ctrl.Rule(service['amazing'] | interior['good'], tip['great'])
rule6 = ctrl.Rule(quota['good'] | interior['good'], tip['great'])
rule7 = ctrl.Rule(service['acceptable'], tip['average'])
rule8 = ctrl.Rule(service['acceptable'] | interior['poor'], tip['bad'])
rule9 = ctrl.Rule(quota['average'] | quality['great'], tip['great'])

tipping_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9])
tipping = ctrl.ControlSystemSimulation(tipping_ctrl)

def callInput(tipping_par):
    def get_input(srv, qlt, intr, qta):
        tipping_par.input['service'] = srv
        tipping_par.input['quality'] = qlt
        tipping_par.input['interior'] = intr
        tipping_par.input['quota'] = qta
    
        tipping_par.compute()
    
        print('Recommended tip', tipping_par.output['tip'])
        tip.view(sim=tipping_par)

    interact_manual(get_input, srv = widgets.FloatSlider(min = 0.0, max = 10.0, step = 0.1, value = 5.0), qlt = widgets.FloatSlider(min = 0.0, max = 10.0, step = 0.1, value = 5.0), intr = widgets.FloatSlider(min = 0.0, max = 10.0, step = 0.1, value = 5.0), qta = widgets.FloatSlider(min = 0.0, max = 10.0, step = 0.1, value = 5.0))

callInput(tipping)
```
